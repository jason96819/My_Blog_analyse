인덱스 6번이 col로 들어가야합니다. skiprows와 인덱싱으로 처리하겠습니다.

```python
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/project/blog/data/22_2.xlsx", skiprows=range(0,6))
df.columns = df.iloc[0]
df = df.iloc[1:]
df.head(10)
```

31번 인덱스가 nan입니다. nan인 행은 지워주겠습니다.

```python
df = df.dropna(how='any')
df.head(10)
```

'작성일' 컬럼을 datetime 객체로 바꾸어주어야합니다. 그러기 위해서 먼저 ". (화)"를 지워줘야합니다.

```python
df['작성일'] = df['작성일'].apply(lambda x: x[:-5])
df.head(10)
```

이제 to_datetime()메서드로 교체해줍시다.

```python
df['작성일'] = pd.to_datetime(df['작성일'], format='%Y.%m.%d')
df.head(10)
```

이제 필요없는 '순위'컬럼은 지우고 '작성일', '제목', '조회수'순으로 컬럼을 정렬하고 작성일 기준으로 정렬하겠습니다.

```python
df = df[['작성일', '제목', '조회수']]
df = df.sort_values(by='작성일', ascending=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

이렇게 22년 2월 데이터를 정제해보았습니다. 함수화 해보겠습니다.

```python
def mkdf(filename):
    df = pd.read_excel(filename, skiprows=range(0,6))
    df.columns = df.iloc[0]
    df = df.iloc[1:]
    df = df.dropna(how='any')
    df['작성일'] = df['작성일'].apply(lambda x: x[:-5])
    df['작성일'] = pd.to_datetime(df['작성일'], format='%Y.%m.%d')
    df = df[['작성일', '제목', '조회수']]
    df = df.sort_values(by='작성일', ascending=True)
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
# 테스트
mkdf("/content/drive/MyDrive/Colab Notebooks/project/blog/data/22_2.xlsx")
```

```python
# /content/drive/MyDrive/Colab Notebooks/project/blog/data/22_{}.xlsx

dum_list = []

for i in range(2, 13):
    dum_list.append(f"/content/drive/MyDrive/Colab Notebooks/project/blog/data/22_{i}.xlsx")
for i in range(1, 13):
    dum_list.append(f"/content/drive/MyDrive/Colab Notebooks/project/blog/data/23_{i}.xlsx")
for i in range(1, 4):
    dum_list.append(f"/content/drive/MyDrive/Colab Notebooks/project/blog/data/24_{i}.xlsx")
```

```python
dum_list
```

```python
for i, filename in enumerate(dum_list):
    globals()[f'df_{i}'] = mkdf(filename)
```

```python
result_df = pd.concat([globals()[f'df_{i}'] for i in range(len(dum_list))], ignore_index=True)
result_df['조회수'] = result_df['조회수'].astype(int)
result_df = result_df.sort_values(by='작성일', ascending=True)
result_df.reset_index(drop=True, inplace=True)
result_df
```

```python
series_data = result_df.groupby(['작성일', '제목'])['조회수'].sum()
series_data
```

```python
df = series_data.reset_index()
df.columns = ['작성일', '제목', '조회수']
df
```

```python
df.to_csv('/content/drive/MyDrive/Colab Notebooks/project/blog/data/myblogdata.csv')
```
